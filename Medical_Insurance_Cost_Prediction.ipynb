```python
# Data Pre-proccessing

# Encoding categorical variables

# Encoding the sex
insurance.replace({'sex':{'male':0,'female':1}},inplace=True)

#encoding smokers
insurance.replace({'smoker':{'yes':0,'no':1}},inplace=True)

#encoding region
insurance.replace({'region':{'southeast':0,'southwest':1,'northwest':2,'northeast':3}},inplace=True)



```

```python
#IQR method for bmi
#Find quartiles
q1=insurance['bmi'].quantile(0.25)
q3=insurance['bmi'].quantile(0.75)
iqr=q3-q1
q1,q3,iqr
```

```python
upper_limit=q3+(1.5*iqr)
lower_limit=q1-(1.5*iqr)
upper_limit,lower_limit
```

```python
sns.boxplot(insurance['bmi'])
```

```python
# find the outliers
insurance.loc[(insurance['bmi'] > upper_limit) | (insurance['bmi'] < lower_limit)]
```

```python
#capping- set the outliers into lower or upper limit
new_insurance=insurance.copy()
new_insurance.loc[new_insurance['bmi']>upper_limit,'bmi']=upper_limit
new_insurance.loc[new_insurance['bmi']<lower_limit,'bmi']=lower_limit
```

```python
upper_limit=q3+(1.5*iqr)
lower_limit=q1-(1.5*iqr)
upper_limit,lower_limit
```

```python
# find the outliers
new_insurance.loc[(new_insurance['charges'] > upper_limit) | (new_insurance['charges'] < lower_limit)]
```

```python
#capping- set the outliers into lower or upper limit

new_insurance.loc[new_insurance['charges']>upper_limit,'charges']=upper_limit
new_insurance.loc[new_insurance['charges']<lower_limit,'charges']=lower_limit
```

```python
new_insurance.head()
```

```python
# splitting the target and features
X=new_insurance.drop(columns='charges',axis=1)
Y=new_insurance['charges']
```

```python
print(X)
```

```python
print(Y)
```

```python
# split the dataset
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Model Training
regressor=LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

```python
#Model evaluation

#make predictions on training data
training_data_prediction=regressor.predict(X_train)
```

```python
#perform R squared
r2_train=metrics.r2_score(Y_train,training_data_prediction)
print('R2 value : ',r2_train)
```

```python
#make predictions on testing data
testing_data_prediction=regressor.predict(X_test)
```

```python
#perform R squared
r2_test=metrics.r2_score(Y_test,testing_data_prediction)
print('R2 value : ',r2_test)
```

```python


# Assuming you have trained your regressor with a DataFrame containing these feature names
feature_names = ['age', 'sex', 'bmi', 'children', 'smoker', 'region']

# Your input data
input_data = (19,1,27.9,0,0,1)

# Change input data into a pandas DataFrame with the same feature names
input_data_df = pd.DataFrame([input_data], columns=feature_names)

# Make the prediction
prediction = regressor.predict(input_data_df)

print(prediction)
print('The Insurance Cost is USD ', prediction[0])
```

```python
# Assuming your model is 'model' and your test data is 'X_test', 'y_test'

residuals = Y_test -testing_data_prediction
print(residuals)
```

```python
# Histogram of residuals
plt.figure(figsize=(8,6))
sns.histplot(residuals, kde=True, bins=30)
plt.title("Histogram of Residuals")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()
```

```python
import scipy.stats as stats

# Q-Q plot
plt.figure(figsize=(8,6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Q-Q Plot of Residuals")
plt.show()
```
